```python
import numpy as np
import pandas as pd
import os
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn import metrics
from sklearn.pipeline import Pipeline
```

# text analyze using frequency (all the word appear in all text except unknown text)

```python

vocab = {}
vocab_word = {}
i = 1


""" 
count all catagory
"""
path = "train/"
for filename in os.listdir(path):
    
    with open(os.path.join(path, filename),'r') as txt:

        x = txt.read().lower().split()   
#         print(x)    
    for word in x:
        if word in vocab:
            continue
        else:
            vocab[word]=i
            vocab_word[i]=word
            i+=1
                

""" 
count catagory's number
"""                   
                    
one = ['1.txt']+[0]*len(vocab)
for filename in os.listdir(path):
    if re.match("CURRENT SMOKER_ID_\d+.txt", filename):
        with open(os.path.join(path, filename),'r') as txt:
            
            x = txt.read().lower().split('\n')
        
 #       print(x)
        for sentance in x:
            words = sentance.split()
            for word in words:
                one[vocab[word]]+=1

for filename in os.listdir(path):
    if re.match("NON-SMOKER_ID_\d+.txt", filename):
        with open(os.path.join(path, filename),'r') as txt:

            x = txt.read().lower().split('\n')
        
        for sentance in x:
            words = sentance.split()
            for word in words:
                one[vocab[word]]+=1

for filename in os.listdir(path):
    if re.match("PAST SMOKER_ID_\d+.txt", filename):
        with open(os.path.join(path, filename),'r') as txt:

            x = txt.read().lower().split('\n')
        
        for sentance in x:
            words = sentance.split()
            for word in words:
                one[vocab[word]]+=1


        
""" 
minus unknown_ID
"""     
        

path = "train/"

                               

for filename in os.listdir(path):
    if re.match("UNKNOWN_ID_\d+.txt", filename):
        with open(os.path.join(path, filename),'r') as txt:

            x = txt.read().lower().split()
        for word in x:

            one[vocab[word]]=0
            
            
            
            

print(one)
# print(vocab_word)
for i in range(len(one)-1):
    if(33>one[i+1]>=10):
#         print(f'{vocab_word[i+1]:{10}} {one[i+1]:{10}} ')
        print('{0:15} {1}'.format(vocab_word[i+1], one[i+1]))
```
